```python
#pyplot API
#Univariate - numerical data
data = {
    "Salary": [50000, 60000, 55000, 70000, 80000, 75000, 90000],
}
df1 = pd.DataFrame(data)
df1.head()
```

```python
#line plot
plt.plot(df1['Salary'],color='blue',marker='o',linestyle='--')
plt.grid()
plt.show()
```

```python
#Histrogram
plt.hist(df1['Salary'])
plt.show()
```

```python
#boxplot
plt.boxplot(df1['Salary'])
plt.show()
```

```python
df1.drop(index=3, inplace=True)  # Drop the row with index 3
df1.shape
```

```python
#Univariate - categorical data
data = {
    "Department": ["HR", "Finance", "IT", "Marketing", "Sales", "IT", "Finance", "HR"],
}
df2 = pd.DataFrame(data)
df = pd.concat([df1, df2], axis=1)
df.head()
```

```python
#pie chart
counts = df2['Department'].value_counts()
counts
```

```python
plt.pie(counts, labels=counts.index, autopct='%1.1f%%',explode=(0, 0.2, 0.3,0,0.1))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#countplot
plt.bar(counts.index, counts.values)
plt.show()
```

```python
data = {
    "Age": [25, 30, 22, 35, 28, 40, 32],
}
df3 = pd.DataFrame(data)
df = pd.concat([df1,df2,df3], axis=1)
df.head()
```

```python
#scatter plot
plt.scatter(df['Age'], df['Salary'], color='green', marker='o')
plt.show()
```

```python
plt.plot(df['Age'], df['Salary'], color='red', marker='o', linestyle='--')
plt.grid()  
plt.show()  
```

```python
hr_salary = df[df['Department'] == 'HR']['Salary']
hr_salary
```

```python
## multivariate - numerical vs categorical
data = {
    "Experience": [2, 5, 3, 7, 4, 10, 6],
}
df4 = pd.DataFrame(data)
df = pd.concat([df1,df2,df3,df4], axis=1)
df.head()
```

```python
##bubble plot
plt.scatter(df['Age'], df['Salary'], s=df['Experience']*50, color ='blue',edgecolors='black')
plt.xlabel('Age')
plt.ylabel('Salary')  
plt.title('Age vs Salary with Experience as Bubble Size')   
plt.grid()  
plt.show()
```

```python
## object oriented API
fig, ax = plt.subplots(1,3, figsize=(15,5))
```

```python
#line plot:

ax[0].plot(df['Age'], df['Salary'], color='red', marker='o', linestyle='--')
ax[0].grid()
ax[0].set_title('line plot')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Salary')

#boxplot:

ax[1].boxplot(df['Salary'])
ax[1].set_title('Boxplot of Salary')
ax[1].set_ylabel('Salary')

# histogram:

ax[2].hist(df['Salary'], bins=5, color='orange', edgecolor='black')
ax[2].set_title('Histogram of Salary')  
ax[2].set_xlabel('Salary')
ax[2].set_ylabel('Frequency')
ax[2].legend(['Salary Distribution'])

plt.tight_layout
plt.show()
```

```python
data = {
    "Year": [2015, 2016, 2017, 2018],
    "Sales": [200, 250, 300, 350],
    "Profit": [20, 30, 50, 70],
    "Expenses": [150, 180, 200, 250]
}
df_financial = pd.DataFrame(data)
df_financial
```

```python
plt.plot(df_financial['Year'], df_financial['Sales'], label='Sales')
plt.plot(df_financial['Year'], df_financial['Profit'], label='Profit')
plt.plot(df_financial['Year'], df_financial['Expenses'], label='Expenses')  
plt.xlabel('Year')
plt.ylabel('Amount in $1000s')
plt.title('Company Financials Over Years')
plt.legend()
plt.grid()
plt.show()
```
