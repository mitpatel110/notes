```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
week_data = [1, 2, 3, 4, 5]
sales_data =[2, 4, 5, 4, 5]

df = pd.DataFrame({
    "Week": week_data,
    "Sales": sales_data
})
df
```

```python
X = df["Week"].values
Y = df["Sales"].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)

print(f"m: {m} ")
print(f"c: {c} ")
```

```python
#plot
max_x = np.max(X) + 1
min_x = np.min(X) - 1

x = np.linspace(min_x, max_x)
y = (m * x) + c

plt.plot(x, y, color='blue', label='Regression Line', linestyle='--')
plt.scatter(X, Y, color='red', label='Scatter Plot') 
plt.xlabel('Week')
plt.ylabel('Sales')
plt.legend(loc = 'best')
plt.grid(True, alpha = 0.3, linestyle = '--')
plt.title('Week vs Sales')
plt.show()
```

```python
#r2
ss_t = 0
ss_r = 0
for i in range(n):
    y_pred = (m * X[i]) + c
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
    r2 = 1 - (ss_r/ss_t)
print(f"R2: {r2}")
```

```python
#validate
from sklearn.linear_model import LinearRegression
X = X.reshape((n, 1))
X
```

```python
reg = LinearRegression()
reg.fit(X, Y)
```

```python
r2 = reg.score(X, Y)
r2
```
